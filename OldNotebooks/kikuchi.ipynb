# Fourier filtering

```python
%matplotlib inline
from scipy import misc
from scipy.ndimage import gaussian_filter

import matplotlib.pyplot as plt
```

```python
ascent = misc.ascent()
```

```python
plt.imshow(ascent)
```

```python
result = gaussian_filter(ascent, sigma=100)
plt.imshow(result)
```

```python
no_background = ascent - result
plt.imshow(no_background)
```
